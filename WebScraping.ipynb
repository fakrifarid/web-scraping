Scrape semua element (judul pekerjaan, nama perusahaan, tanggal job posting) dari website berikut 

https://realpython.github.io/fake-jobs/  menggunakan BeautifulSoup dan simpan data tersebut kedalam csv file.

```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
from requests import get
```

```python
url = "https://realpython.github.io/fake-jobs/"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
```

## Extract Nama Pekerjaan

```python
job = soup.find_all('h2', class_='title is-5')
job
```

```python
list_job = []

for j in job:
    list_job.append(j.text)

list_job
```

## Extract Nama Perusahaan

```python
company = soup.find_all('h3', class_='subtitle is-6 company')
company
```

```python
list_company = []

for c in company:
    list_company.append(c.text)

list_company
```

## Extract Tanggal Publish

```python
date = soup.find_all('time')
date
```

```python
list_date = []

for d in date:
    list_date.append(d.text)

list_date
```

## Membuat dataframe untuk data yang sudah di scarping

```python
df = pd.DataFrame({
                   'job' : list_job,
                   'company' : list_company,
                   'date' : list_date
                   })

df
```

## Tranformasi tambah kolom ID

```python
from uuid import uuid4
import uuid

df['uuid'] = [uuid.uuid4() for _ in range(len(df.index))]
df = df[['uuid','job','company','date']]
df
```

## Masukin data ke CSV

```python
df.to_csv('Web Scraping.csv', index = False)
```
